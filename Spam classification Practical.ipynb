```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(dataset)):
    review = re.sub('[^a-zA-Z0-9]',' ',dataset['message'][i])
    review = review.lower()
    # print(review)
    review = review.split()
    # print(review)
    review = [ps.stem(word) for word in review if not word in stopword_list]
    # print(review)
    review = ' '.join(review)
    # print(review)
    corpus.append(review)
```

```python
corpus
```

#### Independent and Dependent feature

```python
# x = corpus
y = pd.get_dummies(dataset['label'],drop_first=True)
```

```python
y
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size = 0.20, random_state = 0)
```

```python
dataset['label'].value_counts()
```

#### create the BOW

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X_train=cv.fit_transform(X_train).toarray()
```

```python
X_train
```

```python
X_test=cv.transform(X_test).toarray()
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier().fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
